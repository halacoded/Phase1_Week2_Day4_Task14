```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/drive/MyDrive/PHASE1/Week2/Day4/employees.csv")
df
```

###1 Create a line plot showing the trend of salaries over the years employees started working. What trends do you observe in salary changes over the years?

```python
salary_trend = df.groupby("Start Year")["Salary"].mean()
plt.figure(figsize=(15, 5))
plt.plot(salary_trend.index, salary_trend.values, marker="o", color="salmon")
plt.xlabel("Start Year")
plt.ylabel("Average Salary")
plt.title("Trend of Salaries Over the Years Employees Started Working")
plt.show()
#IN 1990 WE CAN OBSERVE AVG SALIERS OS EMPLOYEE WAS HIGHEST
```

###2 Create a scatter plot to visualize the relationship between salary and bonus percentage. Does bonus percentage vary with salary?

```python
plt.figure(figsize=(10, 5))
plt.scatter(df["Salary"], df["Bonus %"], alpha=0.5, color="blue")
plt.xlabel("salary")
plt.ylabel("bonus percentage")
plt.title("relationship between salary and bonus percentage")
plt.grid(True)
plt.show()
#bons  vary  wtih salery
```

###3 Create a bar plot showing the total salary by team. Which team has the highest total salary? What insights can you draw from the salary distribution among teams?

```python
team_salary = df.groupby("Team")["Total Salary"].sum()
plt.figure(figsize=(12, 6))
team_salary.plot(kind="bar", color="skyblue")
plt.xlabel("team")
plt.ylabel("total Salary")
plt.title("total salary for Teams")
plt.show()
#we can see bussies and client and finaancce sectors have hight salery which indicate good dependt for theses skills
```

###4 Create a box plot to compare the salaries of male and female employees. What differences do you notice in salary distributions between genders?

```python
plt.figure(figsize=(8, 6))
df.boxplot(column="Salary", by="Gender", patch_artist=True, grid=False)
plt.xlabel("Gender")
plt.ylabel("Salary")
plt.title("Comparison of Salaries Between Male and Female Employees")
plt.show()
#we can see that they salaries are the same for both gender
```

###5 Create a pie chart that represents the proportion of employees in each team. Which team has the largest representation? What does this suggest about the company's structure?

```python
team_counts = df["Team"].value_counts()
plt.figure(figsize=(10, 6))
team_counts.plot(kind="pie", autopct="%1.1f%%", startangle=140, colors=plt.cm.Paired.colors)
plt.title("Proportion of Employees in Each Team")
plt.show()
```

###6 Create a histogram of the Salary column. What does the distribution of salaries look like? Are salaries normally distributed or do they have multiple peaks?

```python
plt.figure(figsize=(10, 6))
plt.hist(df["Salary"], bins=30, edgecolor="black", alpha=0.7)
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Distribution of Salaries")
plt.show()
#multiple peaks
```

###7 Create a bar plot to compare the average bonus percentage across different teams. Which team has the highest average bonus percentage? What might explain this difference?

```python
team_bonus = df.groupby("Team")["Bonus %"].mean()
plt.figure(figsize=(12, 6))
team_bonus.plot(kind="bar", color="lightgreen")
plt.xlabel("Team")
plt.ylabel("Average Bonus Percentage")
plt.title("Comparison of Average Bonus Percentage Across Teams")
plt.show()
#bussine team has hight avg bonus as they have more emplyee and high salires
```

###8 Create a plot to analyze the salaries of employees in senior management versus non-senior management. Create a suitable visualization. What differences do you observe?

```python
plt.figure(figsize=(8, 6))
df.boxplot(column="Salary", by="Senior Management",grid=False)
plt.xlabel("Senior Management")
plt.ylabel("Salary")
plt.title("Salary Comparison: Senior Management vs. Non-Senior Management")
plt.show()
```

###9 Create a bar chart to show the gender distribution within each team. What patterns do you observe regarding gender representation?

```python
gender_distribution = df.groupby(["Team", "Gender"]).size().unstack()
gender_distribution.plot(kind="bar", figsize=(12, 6))
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.title("Gender Distribution Within Each Team")
#we can see on every depeartment male gender dominate filed except on enegineering where they are equal
```
